```python
def domains(name):
    fir = int(name.split('-')[0][3:])
    sec = int(name.split('-')[1][3:])
    
    if fir >= 27 and fir <= 304: 
        dfir = 'NTD'
    elif fir >= 305 and fir <= 328:
        dfir = 'SD1'
    elif fir >= 335 and fir <= 521:
        dfir = 'RBD'
    elif fir >= 522 and fir <= 672:
        dfir = 'SD2'
    elif fir >= 687 and fir <= 1146:
        dfir = 'S2'
    else:
        dfir = 'Loops'
        
    if sec >= 27 and sec <= 304: 
        dsec = 'NTD'
    elif sec >= 305 and sec <= 328:
        dsec = 'SD1'
    elif sec >= 335 and sec <= 521:
        dsec = 'RBD'
    elif sec >= 522 and sec <= 672:
        dsec = 'SD2'
    elif sec >= 687 and sec <= 1146:
        dsec = 'S2'
    else:
        dsec = 'Loops'
        
    return dfir+'-'+dsec
```

```python
u2d_form_domain = []
for i in u2d_form:
    u2d_form_domain.append(domains(i))
```

```python
u2d_break_domain = []
for i in u2d_break:
    u2d_break_domain.append(domains(i))
```

```python
k = 0
# for fi in {'ASP820-ARG815'}:
for fi in d2u_break:
    dic = d2u_dic.copy()
    o_or_c = 'Opening'
    
    dist = np.zeros([5,400])
    for jj in range(1,6):
        dist[jj-1] = dic[jj][fi][1][:400]
        
    fig = plt.figure(figsize=[4,2.75])
    ax = fig.add_subplot(111, aspect='auto')
    plt.plot(np.arange(400)*0.01, np.mean(dist,axis=0), label=fi, color=sbcolor[fi])
    plt.fill_between(np.arange(400)*0.01, np.min(dist,axis=0), np.max(dist,axis=0), alpha=0.4, 
                     edgecolor=sbcolor[fi], facecolor=sbcolorf[fi], 
                     linewidth=0, linestyle='dashdot', antialiased=True)
    ax.tick_params(labelbottom=True, labeltop=False, labelleft=True, labelright=True,
               bottom=True, top=False, left=True, right=True)
    plt.xlim(0, 4)
    plt.xlabel('Simulation Time (ns)', fontsize=11)
    plt.ylabel('N-O Distance ($\mathrm{\AA}$)', fontsize=11)
    aaa = fi[:3]
    bbb = fi.split('-')[0][3:]
    ccc = fi.split('-')[1][:3]
    ddd = fi.split('-')[1][3:]
    ax.set_title(f'{onekey[aaa]+bbb}-{onekey[ccc]+ddd}' +' in ' + o_or_c +' Transitions', fontsize=12)
    plt.tight_layout()
    plt.savefig(f'ReportPlots/SaltBridgePlot/{o_or_c}_{fi}', dpi=300)
    plt.close()
    
    k += 1
```
